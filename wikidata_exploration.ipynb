```python
filename = 'data/split-0000.json.bz2'

import bz2
import json

def wikidata_parsing(filename):
    with bz2.open(filename, mode='rt', encoding='utf-8') as f:
        f.read(2)
        for line in f:
            try:
                yield json.loads(line.rstrip(',\n'))
            except json.decoder.JSONDecodeError:
                continue

entities = []
for i, data in enumerate(wikidata_parsing(filename)):
    entities.append(data)
    if i == 9999: break
```

```python
def as_json(data, indent=4):
    json_form = json.dumps(data, indent=indent)
    print(json_form)
    return json_form
```
